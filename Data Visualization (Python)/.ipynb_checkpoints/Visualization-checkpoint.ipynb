```python
# general use imports
%pylab inline
import datetime
import numpy as np
import os
import six
import warnings
import matplotlib as mplib
import matplotlib.pyplot as plt # visualization package 1
import seaborn as sns           # visualization package 2
import re

# pandas-related imports
from __future__ import print_function
import pandas as pd
import scipy
import sklearn

# record linkage package
import recordlinkage as rl
from recordlinkage.preprocessing import clean, phonenumbers, phonetic

# CSV file reading-related imports
import csv

# sqlalchemy an psycopg2 are sql connection packages
import sqlalchemy
from sqlalchemy import create_engine

%matplotlib inline

# load dataset
ds = pd.read_csv('./data_frame_1.csv') 
```

```python
# create variable of top 20 organizations with most amount of funds
sns.set_style("whitegrid")
orgfund = ds.groupby('organization_name')['fy_total_cost'].sum().reset_index()
orgfund.columns = ['org','fund']
# Create set of top 20 organizations with most amount of grants
orgfundtop = orgfund.sort_values(by=['fund'],ascending=False).head(20)
orgfundtop
# Graph of top 20 organizations with most amount of grants
sns.set_style("whitegrid")
sns.barplot(x='fund', y='org', data=orgfundtop, palette="PuBuGn_d")
plt.tight_layout()
plt.title('Top 20 organizations: total funding', fontsize=16, fontweight='bold')
plt.xlabel('Total amount funded', horizontalalignment='left', x=-0.5)
plt.ylabel('Organization')
plt.annotate('Source: Federal Reporter Subset', xy=(-0.6,-0.2), xycoords="axes fraction")
plt.savefig('top20fund', bbox_inches='tight')
plt.show()
```
